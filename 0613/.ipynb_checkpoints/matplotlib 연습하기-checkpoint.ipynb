```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [2, 4, 9, 16])
plt.show()
```

- dictionary형 데이터로 x축, y축 지정하기

```python
data_dic = {'data_x':[1, 2, 3, 4, 5], 'data_y':[2, 3, 5, 10, 7]}
plt.plot('data_x', 'data_y', data=data_dic)
plt.show()
```

- 라벨 패딩, 폰트 스타일, 위치 지정

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x-label', labelpad=15)
plt.ylabel('y-label', labelpad=20)
plt.show()
```

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x-label', labelpad=15, fontdict=font1, loc='right')
plt.ylabel('y-label', labelpad=20, fontdict=font2, loc='top')
plt.show()
```

- 범례 추가하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.legend(loc=(0.5, 0.5)) # 그래프의 빈 공간에 자동 배치
# loc로 위치를 조정할 수 있음.
plt.show()
```

- 다중 그래프의 작성

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(ncol=2, fontsize=14) # legend 가로 출력
# 그리드 추가(alpha = 투명도)
plt.grid(True, axis='y', color='red', linestyle='--', alpha = 0.3)
plt.title('title', pad=20)
plt.show()
```

- 막대 그래프와 스타일

```python
import numpy as np
```

```python
x = np.arange(3)
year = ['2021', '2022', '2023']
value = [100, 400, 900]

plt.bar(x, value, color=['r', 'g', 'b'], width=0.4) # bar 그래프
plt.xticks(x, year) # x축 눈금 레이블 표현

plt.show()
```

```python
x = np.arange(3)
year = ['2021', '2022', '2023']
value = [100, 400, 900]

plt.barh(x, value, color='lavender', edgecolor='b', height=0.4) # bar 그래프
plt.yticks(x, year) # 축에 레이블

plt.show()
```

- 산점도 그래프 그리기
    - 산점도 (Scatter plot) : 두 변수의 상관 관계를 직교 좌표계의 평면에 점으로 표현
    - 데이터의 분포도를 확인할 때 사용

```python
n = 50

x = np.random.rand(50) # 0 ~ 1 사이의 난수 50개 발생
y = np.random.rand(50)

area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)
# cmap: 특정 기준으로 도표를 묶음
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

- 파이차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'fig']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

    - 시작 지점과 데이터의 방향도 지정할 수 있다.

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'fig']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
```

        - 차트를 조각낼 때는 explode 사용
        - 도넛 차트를 만들 때는 wedgeprops 속성 사용

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'fig']
explode = [0, 0.10, 0, 0.15]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width':0.7}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,
       explode=explode, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

- 여러 개의 그래프를 하나의 도면에 그리기
    - plt.subplot(row, column, index)

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('1st')

plt.subplot(2, 1, 2)
plt.plot(x1, y1, 'r.-')
plt.title('2nd')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout(h_pad=3) # 두 그래프 사이 패딩

plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'o-')
plt.title('1st')

plt.subplot(1, 2, 2)
plt.plot(x1, y1, 'r.-')
plt.title('2nd')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout(w_pad=3) # 두 그래프 사이 패딩

plt.show()
```

- 박스 플롯
    - 값의 분포도를 시각화

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], positions=[2, 3, 4])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

- 한 축을 공유하는 두 개의 그래프 형성

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])
# 3. 그래프 그리기
fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0, 18) # y축에 최소 / 최대값 제한
ax1.set_xlabel('Year')
ax1.set_ylabel('Price ($)')
ax1.tick_params(axis='both', direction='in') # 파라미터값 설정

ax2 = ax1.twinx() # x축을 공유하는 이중 y축 생성
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0, 18)
ax2.set_ylabel(r'Demand ($\times10^6$)')
ax2.tick_params(axis='y', direction='in')

plt.show()
```
