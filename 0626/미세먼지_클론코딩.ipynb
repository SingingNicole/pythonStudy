```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path="./data/dust.xlsx"
dust = pd.read_excel(file_path)
dust.head()
```

```python
dust.shape
```

```python
dust.info()
```

```python
dust.describe()
```

```python
dust.rename(columns={'날짜':'date', '아황산가스':'so2', '일산화탄소':'co', '오존':'o3', '이산화질소':'no2'}, inplace=True)
dust.head()
```

```python
dust['date'] = dust['date'].str[:11]
dust.head()
```

```python
dust.info()
```

```python
dust['date'] = pd.to_datetime(dust['date'])
dust.dtypes
```

```python
dust['date']
```

```python
dust['year'] = dust['date'].dt.year
dust['month'] = dust['date'].dt.month
dust['day'] = dust['date'].dt.day
dust.columns
```

```python
dust = dust[['date', 'year', 'month', 'day', 'so2', 'co', 'o3', 'no2', 'PM10', 'PM2.5']]
dust.columns
```

```python
dust['month'] = dust['month'].astype(int)
dust['day'] = dust['day'].astype(int)
dust.dtypes
```

```python
dust.shape[0]
```

```python
dust.isnull().sum()
dust
```

```python
dust = dust.fillna(method = 'pad') # 시계열분석이므로 null 값을 이전 시간의 값을 기준으로 채운다.
dust.head()
```

```python
# 이전 값이 없는 경우 NaN을 20으로 치환
dust.fillna(20, inplace=True)
dust.head()
```

```python
dust.isnull().sum()
```

```python
dust.dtypes
```

```python
file_path = './data/weather.xlsx'
weather = pd.read_excel(file_path)
weather.head()
```

```python
weather.info()
```

```python
weather.drop('지점', axis=1, inplace=True)
weather.drop('지점명', axis=1, inplace=True)
weather.head()
```

```python
weather.columns=['date', 'temp', 'wind', 'rain', 'humid']
weather.head()
weather.columns
```

```python
weather['date'] = pd.to_datetime(weather['date']).dt.date
weather['date'] = pd.to_datetime(weather['date'])
weather.info()
weather.head()
```

```python
weather.isnull().sum()
```

```python
weather['rain'].value_counts()
```

```python
# 강수량이 0인 값을 0.01로 변환
weather['rain'] = weather['rain'].replace([0], 0.01)
weather['rain'].value_counts()
```

```python
dust.shape
```

```python
weather.shape
```

```python
weather
```

```python
df = pd.merge(dust, weather, on='date')
df.head()
```

```python
df.corr()
```

```python
corr=  df.corr()
corr
corr['PM10'].sort_values(ascending=False)
```

```python
df.hist(bins=50, figsize=(20, 15))
```

```python
plt.figure(figsize=(15, 10))
dayGraph = sns.barplot(x='day', y='PM10', data=df)
plt.xticks(rotation=0)
plt.show()
```

```python
day = df['day']
plt.hist(day, facecolor='g', alpha=0.5)
plt.title('day frequency')
plt.xlabel('day')
plt.ylabel('freq')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(data=corr, annot=True, fmt='.2f', cmap='hot')
```
